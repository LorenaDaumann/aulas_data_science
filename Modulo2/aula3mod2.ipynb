```python
%matplotlib inline
import matplotlib.pyplot as plt

# Line plot
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.savefig('grafico_linhas.png')
plt.close()

# Bar chart
x = ['Maçãs', 'Bananas', 'Laranjas']
y = [5, 7, 9]

plt.xlabel('Frutas')
plt.bar(x, y)
plt.ylabel('Quantidade')
plt.title('Quantidade de Frutas')

plt.savefig('grafico_barras.png')
```
